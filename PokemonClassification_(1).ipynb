```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread
import random, datetime, os, shutil, math

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
def show_pokemon(pokemon):
  num = len(pokemon)
  if num == 0:
    return None
  rows = int(math.sqrt(num))
  cols = (num+1)//rows
  f, axs = plt.subplots(rows, cols)
  fig = 0
  for b in pokemon:
    img = image.load_img(b)
    row = fig // cols
    col = fig % cols
    axs[row, col].imshow(img)
    fig += 1
  plt.show()
```

```python
dir_name = os.path.join(train_dir,"Pikachu")
all_images = [os.path.join(dir_name, fname) for fname in os.listdir(dir_name)]
show_pokemon(all_images[:6])
```

```python
image_size = (64, 64, 3)
datagen=ImageDataGenerator(rescale = 1./255,
                           shear_range=0.2,
                           zoom_range=0.2,
                           horizontal_flip=True,
                           )
```

```python
training_set=datagen.flow_from_directory(train_dir,
                                         target_size=image_size[:2],
                                         batch_size=32,
                                         class_mode='categorical',
                                         color_mode='rgb'
                                         )
```

```python
validation_set=datagen.flow_from_directory(test_dir,
                                           target_size=image_size[:2],
                                           batch_size=32,
                                           class_mode='categorical',
                                           color_mode='rgb'
                                           )
```

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, GlobalAveragePooling2D
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from keras.models import Sequential
from keras import applications
from keras.utils import plot_model
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=7)
filepath = "model.keras"
ckpt = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
rlp = ReduceLROnPlateau(monitor='loss', patience=3, verbose=1)

```

```python
#defining model
def cnn(image_size, num_classes):
    model = Sequential()

    model.add(Conv2D(64, (5, 5), input_shape=image_size, activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size = (2, 2)))

    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Flatten())

    model.add(Dense(num_classes, activation = 'softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model

neuralnetwork_cnn = cnn(image_size, num_classes)
neuralnetwork_cnn.summary()
plot_model(neuralnetwork_cnn, show_shapes=True)
```

```python
history = neuralnetwork_cnn.fit(
    x=training_set, validation_data=validation_set,
    callbacks=[es, ckpt, rlp], epochs = 20,
)
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
pd.DataFrame(history.history).iloc[:, :-1].plot(ax=ax)
```

```python
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# function to process in batches
def get_predictions_in_batches(model, validation_set, steps_per_epoch):
    y_true_batches = []
    y_pred_batches = []

    for _ in range(steps_per_epoch):
        x_batch, y_batch = next(validation_set)  # Get the next batch from the generator
        y_true_batches.append(y_batch)
        y_pred_prob = model.predict(x_batch)
        y_pred_batches.append(np.argmax(y_pred_prob, axis=1))

    y_true = np.concatenate(y_true_batches, axis=0)
    y_pred = np.concatenate(y_pred_batches, axis=0)

    return y_true, y_pred

# Use the function to get predictions
steps_per_epoch = len(validation_set)  # Or specify steps
y_true, y_pred = get_predictions_in_batches(neuralnetwork_cnn, validation_set, steps_per_epoch=steps_per_epoch)

# Convert y_true to class labels if needed
if y_true.ndim > 1:  # if y_true is one-hot encoded
    y_true = np.argmax(y_true, axis=1)

# Print classification report and accuracy
print("Classification Report:")
print(classification_report(y_true, y_pred))

print("Accuracy Score:")
print(accuracy_score(y_true, y_pred))
```

```python
from IPython.display import display
from ipywidgets import FileUpload, Button, Output
from PIL import Image
import io

# Create an upload button
upload = FileUpload(accept='.jpg,.png', multiple=False)
predict_button = Button(description="Predict")
output = Output()

display(upload, predict_button, output)

def on_predict_button_clicked(b):
    with output:
        output.clear_output()  # Clear previous outputs
        if upload.value:
            # Get uploaded image content
            uploaded_file = list(upload.value.values())[0]
            content = uploaded_file['content']

            # Convert content to an image
            img = Image.open(io.BytesIO(content)).resize((64,64))  # Resize to target size
            img_array = np.array(img) / 255.0  # Normalize if needed
            img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

            # Predict using the model
            y_pred = neuralnetwork_cnn.predict(img_array)
            predicted_label = np.argmax(y_pred)
            predicted_class_name = target_classes[predicted_label]

            # Display prediction
            print("Predicted Class Name:", predicted_class_name)
        else:
            print("Please upload an image file first.")

# Attach the function to the button click event
predict_button.on_click(on_predict_button_clicked)
```
